## data preparation 


```python
data["Description"] = data["Description"].str.strip() # removing white spaces
```

```python
# droping rows that have nan values as their invoices
data["InvoiceNo"].dropna(inplace=True)
```

```python
# now converting the invoice column to string, thats how the algo understand it
data["InvoiceNo"] = data["InvoiceNo"].astype("str")
print(f"value: {data['InvoiceNo'][0]}, datatype: {type(data['InvoiceNo'][0])}")
```

```python
data[data.InvoiceNo.str.contains("C")].head() # this type of invoices mean those are cancelled transaction, therefore they have negative values
```

```python
data = data[~data.InvoiceNo.str.contains("C")] #removing negaative values
```

```python
data.pivot_table(index="InvoiceNo", columns="Description", values="Quantity").fillna(0).head(10)
```

```python
data.groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo").head(10)
```

```python
basket = data.pivot_table(index="InvoiceNo", columns="Description", values="Quantity").fillna(0)
```

```python
# now in order to feed the data into algo we need our data to have only 1 and 0 values
def encode(x):
  if x<=0:
    return 0
  else:
    return 1
```

```python
for i in basket.columns:
  basket[i] = basket[i].apply(encode)

# and we dont need postage column therefore we drop it
basket.drop("POSTAGE", axis=1, inplace=True)
```

```python
basket.head()
```

## Model

```python
# Association rules library
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# now we will find the frequent items and based on that we will try to create association rules
freq_item = apriori(basket, min_support=0.07, use_colnames=True)
```

```python
# now lets generate association rules outta freq_items
a_rules = association_rules(freq_item, metric="lift", min_threshold=1)
```

```python
# printing suport,lift, and confidence
a_rules.head(200)
```

```python
a_rules[(a_rules['support']>0.1) & (a_rules["confidence"]>0.8)]
```
